```python
import json
import numpy as np

with open("/data/long_phan/forecasting_platform/backend/src/benchmarks/output/test_metaculus_full_v14_1_full.json") as file:
    data = json.load(file)

cnt = 0
hard_questions = []
avg_scores = []

cors_model = []
cors_crowd = []
for market in data:
    pred = market['predictions'][0]
    crow_pred = market['crowd_pred'][3][-1]
    origin_crowd_pred = market['crowd_pred'][0][-1]
    resolution = market['resolution']
    

    model_score = (pred - resolution) ** 2

    crowd_score = (crow_pred - resolution) ** 2
    og_crowd_score = (origin_crowd_pred - resolution) ** 2
    cors_model.append(int(pred > .50) == resolution)
    cors_crowd.append(int(crow_pred>.50) == resolution)
    avg_scores.append(model_score)
    if len(market['sources'][0]) == 0:
        cnt += 1
    if model_score - crowd_score > 0.15:
    # if model_score > og_crowd_score:
        question = market['question']
        hard_questions.append(market)
        print("question:", question)
        print("model: ", pred, " | score:", model_score)
        print("crowd: ", crow_pred, " | score:", crowd_score)
        print("=====")

np.mean(cors_crowd), np.mean(cors_model)
```

```python
with open("/data/long_phan/forecasting_platform/backend/src/benchmarks/output/test_full.json") as file:
    data_orig = json.load(file)

with open("/data/long_phan/forecasting_platform/backend/src/benchmarks/output/test_probability_full.json") as file:
    data_pos = json.load(file)

with open("/data/long_phan/forecasting_platform/backend/src/benchmarks/output/test_probability_negation_full.json") as file:
    data_neg = json.load(file)

import numpy as np

scores = []
scores_orgi = []
scores_pos = []
scores_neg = []
for market in data_orig:
    resolution = market['resolution']

    market_pos = [m for m in data_pos if m['question_org'] == market['question']][0]
    market_neg = [m for m in data_neg if m['question_org'] == market_pos['question']][0]

    pred_orig = market['predictions'][0]
    pred_pos = market_pos['predictions'][0]
    pred_neg = 1 - market_neg['predictions'][0]
    # pred_neg = market_neg['predictions'][0]
    pred = np.mean([pred_neg, pred_orig])
    score_org = (pred_orig - resolution) ** 2
    score_pos = (pred_pos - resolution) ** 2
    score_neg = (pred_neg - resolution) ** 2

    score = (pred - resolution) ** 2
    scores.append(score)
    scores_pos.append(score_pos)
    scores_neg.append(score_neg)
    scores_orgi.append(score_org)

    if score_neg - score_pos > 0.2:
        question = market_pos['question']
        print("question pos:", question , " \n| neg: ", market_neg['question'])
        # print("pos: ", pred_pos, " | score:", score_pos)
        # print("neg: ", pred_neg, f" | {market_neg['predictions'][0]} | score:", score_neg)
        # print("=====")

np.mean(scores), np.mean(scores_orgi), np.mean(scores_pos), np.mean(scores_neg)
```

```python
import json
import numpy as np

file_name = "test_metaculus_full_v14_1"
# file_name = "test_metaculus_full_v15_gpt4o-08"

avg_brier = []
avg_acc = []
avg_confidence = []  # List to store average confidence for each run

for i in range(0, 9):
    with open(f"stagging_output/{file_name}_{i}.json") as file:
        data = json.load(file)

    cors_crowd = []
    cors_model = []
    avg_scores = []
    confidences = []  # List to store confidences

    for market in data:
        pred = market['predictions'][0]
        crow_pred = market['crowd_pred'][3][-1]
        resolution = market['resolution']

        score = (pred - resolution) ** 2

        cors_model.append(int(pred > 0.50) == resolution)
        cors_crowd.append(int(crow_pred > 0.50) == resolution)
        avg_scores.append(score)
        
        # Calculate confidence using max(pred, 1-pred)
        confidence = max(pred, 1 - pred)
        confidences.append(confidence)

    assert len(avg_scores) == 177
    avg_brier.append(np.mean(avg_scores))
    avg_acc.append(np.mean(cors_model))
    avg_confidence.append(np.mean(confidences))  # Store average confidence for this run
    print(f"Run {i+1}: Brier Score: {np.mean(avg_scores):.4f}, Model Accuracy: {np.mean(cors_model):.4f}, "
          f"Crowd Accuracy: {np.mean(cors_crowd):.4f}, Avg Confidence: {np.mean(confidences):.4f}, "
          f"Number of predictions: {len(cors_model)}")

print(f"\nOverall Results:")
print(f"Accuracy: {np.mean(avg_acc) * 100:.2f}% ± {np.std(avg_acc) * 100:.2f}")
print(f"Brier Score: {np.mean(avg_brier):.4f} ± {np.std(avg_brier):.4f}")
print(f"Average Confidence: {np.mean(avg_confidence):.4f} ± {np.std(avg_confidence):.4f}")
```
