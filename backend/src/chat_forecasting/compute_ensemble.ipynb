```python
import json
```

```python
files_name = ["test_adam_time.json", "test_adam_time_0.json", "test_adam_time_1.json"]
import numpy as np
datas = []

for file_name in files_name:
    with open(f"/data/long_phan/forecasting_platform/backend/src/benchmarks/output/{file_name}") as file:
        d = json.load(file)
        datas.append(d)

avg_score = []
for m1,m2,m3 in zip(datas[0],datas[1],datas[2]):
    m2 = [m for m in datas[1] if m['id'] == m1['id']][0]
    m3 = [m for m in datas[2] if m['id'] == m1['id']][0]

    assert m1['id'] == m2['id'] == m3['id']

    pred1 = m1['predictions'][0]
    pred2 = m2['predictions'][0]
    pred3 = m3['predictions'][0]
    pred = np.mean([pred1,pred2,pred3])
    score = (pred - m1['resolution']) ** 2
    avg_score.append(score)
np.mean(avg_score)
```
